```python
import pandas as pd
import requests
import numpy as np


NPdf = pd.read_csv('neighbourhood-profiles-2016-csv.csv', index_col= "_id")

#Creating Neighbourhoood Profile dataframe
#Importing and reading CSV file
```

NPdf.head()

```python
NPdf.reset_index(drop=True, inplace=True)
NPdf.set_index("Category", inplace=True)
NPdf.head()

#Data cleaning
#Resetting index to narrow down unwanted columns + rows
```

```python
NPdf = NPdf.drop(['Aboriginal peoples'])
NPdf = NPdf.drop(['Education']) 
NPdf = NPdf.drop(['Families, households and marital status']) 
NPdf = NPdf.drop(['Housing']) 
NPdf = NPdf.drop(['Immigration and citizenship']) 
NPdf = NPdf.drop(['Journey to work']) 
NPdf = NPdf.drop(['Labour']) 
NPdf = NPdf.drop(['Language']) 
NPdf = NPdf.drop(['Language of work']) 
NPdf = NPdf.drop(['Mobility']) 
NPdf = NPdf.drop(['Neighbourhood Information']) 
NPdf = NPdf.drop(['Population']) 
NPdf = NPdf.drop(['Visible minority'])

#Dropping all the unwanted columns 
```

```python
NPdf.reset_index(drop=True, inplace=True)
NPdf.set_index("Topic", inplace=True)
NPdf.head()

#Resetting the index as "Topic" to further clean dataset
```

```python
NPdf = NPdf.drop(['Income of individuals in 2015', 'Income of economic families in 2015', 'Income sources', 'Income taxes', 'Low income in 2015'])
#Dropping more columns
```

```python
NPdf.reset_index(drop=True, inplace=True)
NPdf = NPdf.drop(["Data Source"], axis=1)
NPdf = NPdf.drop_duplicates(subset = "Characteristic", keep = "first")

#Resetting the index 
#Removing duplicate values under "Characteristic" column
```

```python
NPdf.set_index("Characteristic", inplace=True)
NPdf.info()

#Setting index column as "Characteristic"
```

```python
NPdf = NPdf.dropna()

#Dropping all null values
```

```python
NPdf = NPdf.drop(['Total - Income statistics in 2015 for private households by household size - 100% data', '    Total - Income statistics in 2015 for one-person private households - 100% data'])

#Dropping further columns to filter out specific income related data values
```

```python
NPdf = NPdf.drop(['    Total - Income statistics in 2015 for two-or-more-person private households - 100% data', 'Total - Income statistics in 2015for private households by household size - 25% sampledata'])

#Dropping further columns to filter out specific income related data values
```

```python
NPdf = NPdf.drop(['  Average after-tax income of households in 2015 ($)','    Total - Income statistics in 2015 for one-person private households - 25% sample data'])

#Dropping further columns to filter out specific income related data values
```

```python
NPdf= NPdf.drop(['Total - Household total income groups in 2015 for private households - 100% data','Total - Household after-tax income groups in 2015 for private households - 100% data', '    Total - Income statistics in 2015 for two-or-more-person private households - 25% sample data'])

#Dropping further columns to filter out specific income related data values
```
