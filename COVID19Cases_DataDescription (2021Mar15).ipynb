```python
import pandas as pd
import requests
import numpy as np


COVIDdf = pd.read_csv('COVID19 cases.csv')
#Creating COVIDdf dataframe
#Reading COVID19 cases csv file and adding into dataframe
```

```python
COVIDdf.dtypes
#Identifying data types of each value in dataset
```

```python
COVIDdf.info()
#providing quantitatvie information of dataset and data type
```

```python
COVIDdf = COVIDdf.drop(["Outbreak Associated", "FSA", "Source of Infection", "Classification", "Outcome", 
                        "Currently Hospitalized","Currently in ICU","Currently Intubated", "Ever Hospitalized","Ever in ICU",
                        "Ever Intubated"], axis =1)

#Dropping unwanted columns
```

```python
print(COVIDdf.isnull().sum())

#Checking dataset for any null values
```

```python
COVIDdf = COVIDdf.dropna()
COVIDdf.count()

#Dropping null values to clean dataset
```

```python
print(COVIDdf["Neighbourhood Name"].value_counts())
COVIDdf["Neighbourhood Name"].value_counts().plot(kind = 'barh')

#finding value count of each neighbourhood
#plotting as bar graph
```

```python
print(COVIDdf["Age Group"].value_counts())
COVIDdf["Age Group"].value_counts().plot(kind = 'barh')

#finding value count of each age group
#plotting as bar graph
```

```python
print(COVIDdf["Client Gender"].value_counts())
COVIDdf["Client Gender"].value_counts().plot(kind = 'barh')

#finding value count of each gender group
#plotting as bar graph
```

```python
agesexTable = COVIDdf.groupby(["Age Group", "Client Gender"])["Client Gender"].value_counts()
agesexTable

#finding the gender group breakdown within each age
```

```python
agesexTable.plot(kind='barh')
```

```python
print(COVIDdf["Reported Date"].value_counts())
COVIDdf["Reported Date"].value_counts().plot(kind = 'barh')

#finding the number of cases from the time of COVID-19 in Toronto
```
